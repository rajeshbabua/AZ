```python
js_n= pd.json_normalize(j_son.get('results'))
```

```python
js_n.info()
```

```python
js_n
```

```python
js_n["openfda.route"]
```

```python
###removing all the null values in openfda.route column
js_no_null=js_n[~js_n['openfda.route'].isnull()]
```

```python
js_no_null["openfda.route"]
```

```python
###extracting only year from date###
js_no_null['year'] = pd.DatetimeIndex(js_no_null['effective_time']).year
```

```python
js_no_null["year"]
```

```python
js_no_null[["year","spl_product_data_elements","openfda.generic_name","openfda.route"]]
```

```python
js_no_null["spl_product_data_elements"][2]
```

```python
####counting number of ingredients###
js_no_null['number_of_ingredients'] = js_no_null['spl_product_data_elements'].apply(lambda x: len(x[0].split(' ')))
```

```python
js_no_null[["year","spl_product_data_elements","openfda.generic_name","openfda.route",'number_of_ingredients']]
```

```python
(js_no_null["year"])
```

```python
js_no_null[["year","openfda.route","number_of_ingredients"]]
```

```python
###creating new column with all the route methods######
js_no_null['route']= js_no_null['openfda.route'].apply(lambda x:(x[0]))
```

```python
js_no_null= js_no_null[["year","route","number_of_ingredients"]]
```

```python
js_no_null
```

```python
js_n_PART_B= js_no_null.groupby(["year","route"], as_index=False).agg({"number_of_ingredients":'mean'})
```

```python
js_n_PART_B.columns
```

```python
js_n_PART_B.columns=["year","Route","Avg. no of ingredients"]
```
