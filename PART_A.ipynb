```python
###As total no of results are only 36. limit is given as 36#####
url_limit="https://api.fda.gov/drug/label.json?search=openfda.manufacturer_name:"+Company+"&limit=36"
```

```python
r_limit=requests.get(url_limit)
```

```python
##converting to Json format 
js=r_limit.json()
```

```python
###flatening the json format###
js_n= pd.json_normalize(js.get('results'))
```

```python
js_n.info()
```

```python
js_n
```

```python
js_n['effective_time']
```

```python
####Extracting only year from the date############
js_n['year'] = pd.DatetimeIndex(js_n['effective_time']).year
```

```python
js_n["year"]
```

```python
js_n[["year","spl_product_data_elements","openfda.generic_name"]]
```

```python
js_n["spl_product_data_elements"][0]
```

```python
type(js_n["openfda.generic_name"])
```

```python
#####counting the ingredients############
js_n['number_of_ingredients'] = js_n['spl_product_data_elements'].apply(lambda x: len(x[0].split(',')))
```

```python
js_n[["year","spl_product_data_elements","openfda.generic_name",'number_of_ingredients']]
```

```python
js_n["year"].unique()
```

```python
js_n_PART_A= js_n.groupby("year", as_index=False).agg({"openfda.generic_name":'sum',"number_of_ingredients":'mean'})
```

```python
js_n_PART_A
```

```python
js_n_PART_A.columns
```

```python
js_n_PART_A.columns=["year","Drugs_names","Avg. no of ingredients"]
```

```python
js_n_PART_A
```

```python
js_n_PART_A.plot(x='year', y='Avg. no of ingredients',title='Average no of ingredients used by AstraZeneca per year')
```
